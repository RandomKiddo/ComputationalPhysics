### Dependencies Required
- pylab
- numpy

````python
from pylab import plot, show
from numpy import linspace
````

### Graphing a combination of x and y values

````python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 5, 3, 7, 11, 10, 4]
plot(x, y)
show()
````

##### Explanation

The variable `x` defines the x-axis values. This can be represented as a list or tuple. Floats or ints are accepted. If no x variable is entered, each y-value is assumed to match with the next available number starting at 0 (i.e. first y-value is at x=0, second is at x=1, etc.).

The variable `y` defines the y-axis values. This can also be represented as a list or tuple. Floats or ints are also accepted.

The lenghts of `x` and `y` must be the same for the interpreter to understand and for the graph to show.

`plot` plots the graph and `show` displays it

### Linspace Usage

`linspace` is part of the numpy package, and returns evenly spaced numbers over an interval specified by the user

The function does take many parameters [[More Reading]](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html), but the three most important ones for the purposes of graphing are `start`, `stop`, and `num`.

`start` specifies the starting value of the sequence, aka the initial `x` value

`stop` specifies the end value of the sequence (providing that another paramter `endpoint` is set to `True`, which it is by default). 

`num` gives the number of samples to generate (defaults to 50). Parameter must be non-negative. 

````python
linspace(2.0, 3.0, num=5)
````

##### Linspace in Context of Graphing

In order to graph, we can take a range of `x` values, use them to create a range of `y` values, and then graph them side-by-side

````python
from numpy import sin
x = linspace(0, 7, 100) # define a linspace from 0-7 with 100 elements
y = sin(x) # take the sin of all the values in x
plot(x, y)
show()
````

### Loading Graph Values From a Text File

By importing the `loadtxt` function from numpy, values for the graph can be used like the following:

```py
from numpy import loadtxt
from pylab import plot, show

data = loadtxt('values.txt', float)
x = data[:,0]
y = data[:,1]
plot(x, y)
show()
```

This is assumming the `values.txt` file is set up like the following:
```
0   123
1   456
2   789
3   012
```

### Spacing, Labels, Styling

The pylab library has a lot of functionality regarding styling a graph

In this section, we will go over the following:
- Adding spacing to the bounds of a graph
- Adding labels to the axes of the graph
- Creating dashed, dotted, and other lines
- Working with the encoded-string for controlling line colors and its style
- Graphing plots side-by-side

````python
from pylab import ylim # ylim helps create a y-limit that adds spacing to the graph
x = linspace(0, 7, 100)
y = sin(x)
plot(x, y)
ylim(-1.5, 1.5) # sets the y range from -1.5 to 1.5
show()
````

````python
from pylab import xlabel, ylabel
x = linspace(0, 7, 100)
y = sin(x)
plot(x, y)
ylim(-1.5, 1.5)
xlabel('radian angle measure') # sets label of the x-axis to 'radian angle measure'
ylabel('sin(x)') # sets label of the y-axis to 'sin(x)'
show()
````

````python
x = linspace(0, 7, 100)
y = sin(x)
plot(x, y, 'g--') # Third argument controls line style and color, we will cover this soon
show() # shows a green dashed graph
````

````python
from numpy import cos
x = linspace(0, 7, 100)
y1 = sin(x)
y2 = cos(x)
plot(x, y1, 'k-') # graphs sin(x) as a black solid line
plot(x, y2, 'k--') # graphs cos(x) as a black dashed line
show()
````

##### Working With the Style String Argument

The third argument of the `plot` function specifies the style of the plot being graphed.

The letter of the string represents the plot's color. Allowed letters are the following:
`r` --> red <br>
`g` --> green <br>
`b` --> blue <br>
`c` --> cyan <br>
`m` --> magenta <br>
`y` --> yellow <br>
`k` --> black <br>
`w` --> white <br>

The next set of characters following the color letter represents the style of the graph. Allowed character strings are the following:
`-` --> solid line <br>
`--` --> dashed line <br>
`o` --> circles <br>
`s` --> squares <br>
`.` --> thin dots <br>

Examples:
`r--` --> dashed red line <br>
`co` --> cyan line with circle points <br>
`ws` --> white line with square points <br>
`m-` --> solid magenta line <br>
`g.` --> green thin dots
