```python
fig = plot.figure()
octant = plot.axes(projection='3d')
```

### Points and Lines in Space

To plot a line in space, we just need to set up three line equations for the axes and use the `plot3D` function to plot.

```python
octant = plot.axes(projection='3d') # axes must be redefined after displaying in order for plot to show correctly
x = linspace(0, 15, 1000)
y = sin(x)
z = cos(x)
octant.plot3D(x, y, z, 'red') # fourth parameter indicates color
```

To plot sample data, we use the `scatter3D` function:

```python
octant = plot.axes(projection='3d')
# The next three lines are just generating random data
x_data = 15 * np.random.random(100)
y_data = sin(x_data) + 0.1 * np.random.randn(100)
z_data = cos(x_data) + 0.1 * np.random.randn(100)
octant.scatter3D(x_data, y_data, z_data, color='green')
```

Plotting a line and scatter together:

```python
octant = plot.axes(projection='3d') 
x = linspace(0, 15, 1000)
y = sin(x)
z = cos(x)
octant.plot3D(x, y, z, 'gray')
x_data = 15 * np.random.random(100)
y_data = sin(x_data) + 0.1 * np.random.randn(100)
z_data = cos(x_data) + 0.1 * np.random.randn(100)
octant.scatter3D(x_data, y_data, z_data, color='red')
```

### Contour Plots

Contour plots can be made using the `contour3D` function. Let's define a function `f(x, y)` as `cos(x+y)`

The numpy method `meshgrid` will be used to aid in this process. The function returns coordinate matrices for coordinate vectors --> [More Reading](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html)

```python
def f(x, y):
    return cos(x+y)
x = linspace(-10, 10, 100)
y = linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
octant = plot.axes(projection='3d')
octant.contour3D(X, Y, Z, levels=50, cmap='binary') # levels is the number and positions of countor lines and regions, cmap is color map, which is binary in this case
octant.set_xlabel('x') # Sets x-axis label to x
octant.set_ylabel('y') # Sets y-axis label to y
octant.set_zlabel('z') # Sets z-axis label to z
```

To change the viewing angle, we can use the `view_init` function. The function takes elevation and azimuthal angle values. 

```python
def f(x, y):
    return cos(x+y)
x = linspace(-10, 10, 100)
y = linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
octant = plot.axes(projection='3d')
octant.contour3D(X, Y, Z, levels=50, cmap='binary') # levels is the number and positions of countor lines and regions, cmap is color map, which is binary in this case
octant.set_xlabel('x') # Sets x-axis label to x
octant.set_ylabel('y') # Sets y-axis label to y
octant.set_zlabel('z') # Sets z-axis label to z
octant.view_init(60, 30) # elecation angle = 60, azimuthal angle = 30
```

### Wireframes and Surface Plots 

To plot wireframes, we use the `plot_wireframe` function and to plot surfaces we use the `plot_surface` function. Pretty straightforward right?

```python
octant = plot.axes(projection='3d')
octant.plot_wireframe(X, Y, Z, color='red')
octant.set_title('Wireframe') # Sets the plot title to 'Wireframe'
octant.view_init(70, 5) # Angle adjusted for better viewing
```

```python
octant = plot.axes(projection='3d')
octant.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
octant.set_title('Surface')
octant.view_init(60, 30)
```

Zooming in on both of them, we can really see the wireframe and surface patterns:

```python
octant = plot.axes(projection='3d')
octant.plot_wireframe(X, Y, Z, color='red')
octant.set_title('Wireframe') # Sets the plot title to 'Wireframe'
octant.view_init(70, 5) # Angle adjusted for better viewing
octant.margins(x=0, y=0, z=.75) # Adjusting margins to see pattern
```

```python
octant = plot.axes(projection='3d')
octant.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
octant.set_title('Surface')
octant.view_init(60, 30)
octant.margins(x=.1, y=.1, z=.75)
```

### Triangulation

To end this very high-level section, we will talk about triangulating surfaces. That is, finding a surface through connecting triangles. 

Let's say we have the following randomly generated polar grid values:

```python
r = 3 * np.random.random(1000)
theta = 2 * np.pi * np.random.random(1000)
x = np.ravel(r * cos(theta))
y = np.ravel(r * sin(theta))
z = f(x, y)
octant = plot.axes(projection='3d')
octant.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.6)
```

We can use the `plot_trisurf` function to create a triangulated estimation of the surface we're working with. Note that in this case, `x`, `y`, and `z` are one-dimensional arrays. Not two-dimensional like in the previous plotting examples of `X`, `Y`, and `Z`.

```python
octant = plot.axes(projection='3d')
octant.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
```

The surface produced is not as clean as the standard known grid values plot, but is a good estimation and can really form some very accurate and interesting 3D plots
